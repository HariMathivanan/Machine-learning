```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler

x = np.array([1, 2, 4, 6, 5])
y = np.array([1, 3, 4, 3, 7])


x_reshaped = x.reshape(-1, 1)

print("Shape of reshaped x:", x_reshaped.shape)
print("Shape of y:", y.shape)

model = LinearRegression()
model.fit(x_reshaped, y)
print(model.predict([[8]]))
y_pred = model.predict(x_reshaped) 

plt.scatter(x, y, color='r', marker='o', label='Actual data') 
plt.plot(x, y_pred, color='b', linewidth=2, label='Linear Regression prediction') 
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.legend()
plt.show()

print("Linear Regression Coefficients:", model.coef_)
print("Linear Regression Intercept:", model.intercept_)

scaler = StandardScaler()

x_stand = scaler.fit_transform(x.astype(float).reshape(-1, 1)) # Reshape before scaling


sgdr = SGDRegressor(penalty='l2', alpha=0.15, max_iter=100)
sgdr.fit(x_stand, y)
print(sgdr.predict([[10]]))
y_pred2 = sgdr.predict(x_stand)


plt.scatter(x_stand, y, color='r', marker='o', label='Actual data (standardized x)')
plt.plot(x_stand, y_pred2, color='b', linewidth=2, label='SGDRegressor prediction')
plt.xlabel('x_normalization')
plt.ylabel('y')
plt.title("SGD Regressor with Standardized X")
plt.legend()
plt.show()

print("SGD Regressor Coefficients:", sgdr.coef_)
print("SGD Regressor Intercept:", sgdr.intercept_)
```
